```python
import pandas as pd
import re
```

```python
price_df = pd.read_csv('price_df.csv')
```

```python
btc_price_df = price_df[price_df['currency'] == 'btc']
```

```python
btc_price_df = btc_price_df[['date', 'last']]
```

```python
btc_price_df = btc_price_df.dropna()
```

```python
btc_price_df = btc_price_df.reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python
import datetime
```

```python
x = [datetime.datetime.strptime(i, '%Y-%m-%d %H:%M:%S') for i in btc_price_df['date']]
y = btc_price_df['last']
```

```python
plt.plot(x, y)
plt.gcf().autofmt_xdate()
plt.show()
```

```python
temp = []
for i in range(len(btc_price_df['last'])-1):
    if btc_price_df['last'][i] < btc_price_df['last'][i+1]:
        temp.append(1)
    elif btc_price_df['last'][i] == btc_price_df['last'][i+1]:
        temp.append(1)
    else:
        temp.append(0)
temp.append(1)
```
